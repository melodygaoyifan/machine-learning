### Audio Cleaning and augment

```python
# Define preprocessing functions
def resample_audio(waveform, original_sample_rate, target_sample_rate=16000):
    if original_sample_rate != target_sample_rate:
        resampler = torchaudio.transforms.Resample(orig_freq=original_sample_rate, new_freq=target_sample_rate)
        waveform = resampler(waveform)
    return waveform

def normalize_audio(waveform):
    return waveform / waveform.abs().max()

def remove_silence(waveform, sample_rate, threshold=20, min_silence_duration=0.3):
    if waveform.shape[0] > 1:
        waveform = waveform.mean(dim=0, keepdim=True)

    threshold = threshold / 100.0 * waveform.max()
    mask = (waveform.abs() > threshold).float()
    start = (mask[0] == 1).nonzero(as_tuple=True)[0][0]
    end = (mask[0] == 1).nonzero(as_tuple=True)[0][-1]
    waveform = waveform[:, start:end]
    return waveform

def reduce_noise(waveform, sample_rate):
    waveform_np = waveform.detach().numpy()
    reduced_noise = nr.reduce_noise(y=waveform_np, sr=sample_rate)
    reduced_noise_tensor = torch.tensor(reduced_noise)
    return reduced_noise_tensor

def time_shift(waveform, shift_limit=0.2):
    shift = int(random.uniform(-shift_limit, shift_limit) * waveform.size(1))
    return waveform.roll(shift)

def pitch_shift(waveform, sample_rate, n_steps):
    return transforms.PitchShift(sample_rate, n_steps=n_steps)(waveform)

def time_stretch(waveform, rate):
    window = torch.hann_window(400)
    stft = torch.stft(waveform, n_fft=400, hop_length=160, win_length=400, window=window, return_complex=True)
    stretched_stft = transforms.TimeStretch()(stft, rate)
    stretched_waveform = torch.istft(stretched_stft, n_fft=400, hop_length=160, win_length=400, window=window)
    return stretched_waveform

def add_noise(waveform, noise_level=0.005):
    noise = torch.randn_like(waveform) * noise_level
    return waveform + noise

def preprocess_audio(file_path, sample_rate=16000, augment=False):
    try:
        waveform, original_sample_rate = torchaudio.load(file_path)
    except RuntimeError as e:
        print(f"Error loading {file_path}: {e}")
        return None
    waveform = resample_audio(waveform, original_sample_rate, sample_rate)
    waveform = normalize_audio(waveform)
    waveform = remove_silence(waveform, sample_rate)
    waveform = reduce_noise(waveform, sample_rate)

    if augment:
        if random.random() < 0.5:
            waveform = time_shift(waveform)
        if random.random() < 0.5:
            waveform = pitch_shift(waveform, sample_rate, n_steps=random.randint(-5, 5))
        if random.random() < 0.5:
            waveform = time_stretch(waveform, rate=random.uniform(0.8, 1.2))
        if random.random() < 0.5:
            waveform = add_noise(waveform)

    # Ensure waveform is 1D
    waveform = waveform.squeeze()

    return waveform

```

### Filter audio data based on SNR and ZCR

```python
def load_audio(file_path):
    waveform, sample_rate = torchaudio.load(file_path)
    return waveform, sample_rate
def calculate_snr(waveform):
    signal_power = np.mean(waveform.numpy() ** 2)
    noise_power = np.var(waveform.numpy())
    snr = 10 * np.log10(signal_power / noise_power)
    return snr

def calculate_zcr(waveform):
    zero_crossings = np.where(np.diff(np.sign(waveform.numpy())))[0]
    zcr = len(zero_crossings) / len(waveform.numpy())
    return zcr


def filter_and_preprocess_files(directory, output_directory, min_snr, max_snr, min_zcr, max_zcr, augment=False):
    filtered_files = []
    for filename in os.listdir(directory):
        if filename.endswith(".wav"):
            file_path = os.path.join(directory, filename)
            try:
                waveform, sample_rate = load_audio(file_path)
                snr = calculate_snr(waveform)
                zcr = calculate_zcr(waveform)

                if min_snr <= snr <= max_snr and min_zcr <= zcr <= max_zcr:
                    processed_waveform = preprocess_audio(file_path, augment=augment)
                    if processed_waveform is not None:
                        output_file_path = os.path.join(output_directory, filename)
                        torchaudio.save(output_file_path, processed_waveform.unsqueeze(0), sample_rate)
                        filtered_files.append(output_file_path)
            except Exception as e:
                print(f"Error processing file {file_path}: {e}")
    return filtered_files

# Example usage
directory = "/content/GenNote/Wav_Recordings"
output_directory = 'cleaned_wav'
os.makedirs(output_directory, exist_ok=True)

min_snr = -1  # minimum signal-to-noise ratio
max_snr = 1  # maximum signal-to-noise ratio
min_zcr = 0.4 * 1e6  # minimum zero-crossing rate
max_zcr = 1.8 * 1e6  # maximum zero-crossing rate

filtered_files = filter_and_preprocess_files(directory, output_directory, min_snr, max_snr, min_zcr, max_zcr, augment=True)
print("Filtered and processed files:", filtered_files)

# Plotting the distributions of SNR and ZCR
snrs = []
zcrs = []
for wav_file in os.listdir(directory):
    if wav_file.endswith(".wav"):
        file_path = os.path.join(directory, wav_file)
        try:
            waveform, sample_rate = load_audio(file_path)
            snrs.append(calculate_snr(waveform))
            zcrs.append(calculate_zcr(waveform))
        except Exception as e:
            print(f"Error processing file {file_path}: {e}")

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
sns.histplot(snrs, bins=30, kde=True)
plt.title("Distribution of Signal-to-Noise Ratio (SNR)")
plt.xlabel("SNR (dB)")
plt.ylabel("Frequency")

plt.subplot(2, 1, 2)
sns.histplot(zcrs, bins=30, kde=True)
plt.title("Distribution of Zero-Crossing Rate (ZCR)")
plt.xlabel("ZCR")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import os
import random
import torchaudio
import matplotlib.pyplot as plt

def plot_waveform(waveform, sample_rate, title="Waveform"):
    num_channels, num_frames = waveform.shape
    time_axis = torch.arange(0, num_frames) / sample_rate

    figure, axes = plt.subplots(num_channels, 1, figsize=(15, 4))
    if num_channels == 1:
        axes = [axes]
    for c in range(num_channels):
        axes[c].plot(time_axis.numpy(), waveform[c].numpy(), linewidth=1)
        axes[c].grid(True)
        axes[c].set_title(f'{title} (Channel {c+1})')
        axes[c].set_xlabel('Time (s)')
        axes[c].set_ylabel('Amplitude')
    plt.tight_layout()
    plt.show()

def plot_spectrogram(waveform, sample_rate, title="Spectrogram"):
    spectrogram = torchaudio.transforms.MelSpectrogram(sample_rate=sample_rate)(waveform)
    spectrogram_db = torchaudio.transforms.AmplitudeToDB()(spectrogram)

    figure, ax = plt.subplots(figsize=(15, 4))
    ax.imshow(spectrogram_db[0].numpy(), origin='lower', aspect='auto')
    ax.set_title(title)
    ax.set_xlabel('Time (frames)')
    ax.set_ylabel('Frequency (bins)')
    plt.colorbar(ax.imshow(spectrogram_db[0].numpy(), origin='lower', aspect='auto'), ax=ax)
    plt.show()

def main(directory, num_files=3):
    files = [f for f in os.listdir(directory) if f.endswith(".wav")]

    for i, file_name in enumerate(files):
        file_path = os.path.join(directory, file_name)
        waveform, sample_rate = torchaudio.load(file_path)

        plot_waveform(waveform, sample_rate, title=f"Waveform of {file_name}")
        plot_spectrogram(waveform, sample_rate, title=f"Spectrogram of {file_name}")

if __name__ == "__main__":
    directory = "/content/clean_wav"
    main(directory)
```

```python
!pip install torchaudio

import os
import torchaudio
import torch

# Directory containing audio files
audio_dir = "/content/GenNote/Wav_Recordings"

# Function to load and resample audio
def load_and_resample_audio(file_path, target_sample_rate=16000):
    try:
        waveform, sample_rate = torchaudio.load(file_path)
        if sample_rate != target_sample_rate:
            waveform = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=target_sample_rate)(waveform)
        return waveform.squeeze(), target_sample_rate
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        return None, None

# Function to extract MFCC features
def extract_mfcc(audio, sample_rate, n_mfcc, n_fft, hop_length, n_mels):
    mfcc_transform = torchaudio.transforms.MFCC(
        sample_rate=sample_rate,
        n_mfcc=n_mfcc,
        melkwargs={
            'n_mels': n_mels,
            'n_fft': n_fft,
            'hop_length': hop_length
        }
    )
    return mfcc_transform(audio)

# Function to check for non-finite elements
def has_non_finite_elements(tensor):
    return not torch.isfinite(tensor).all()

# Hyperparameter grids
n_mfcc_options = [13, 20, 40]
n_fft_options = [400, 512, 1024]
hop_length_options = [160, 200, 400]
n_mels_options = [23, 40, 80]

# Lists to store the results
best_params = None
best_score = float('inf')

# Process each file in the directory
audio_files = [os.path.join(audio_dir, f) for f in os.listdir(audio_dir) if f.endswith(".wav")]

# Perform grid search
for n_mfcc in n_mfcc_options:
    for n_fft in n_fft_options:
        for hop_length in hop_length_options:
            for n_mels in n_mels_options:
                print(f"Testing parameters: n_mfcc={n_mfcc}, n_fft={n_fft}, hop_length={hop_length}, n_mels={n_mels}")

                total_files = 0
                files_without_non_finite = 0
                for file_path in audio_files:
                    audio, sample_rate = load_and_resample_audio(file_path)
                    if audio is not None:
                        mfcc = extract_mfcc(audio, sample_rate, n_mfcc, n_fft, hop_length, n_mels)
                        total_files += 1
                        if not has_non_finite_elements(mfcc):
                            files_without_non_finite += 1

                # Calculate the score as the percentage of files without non-finite elements
                score = files_without_non_finite / total_files
                print(f"Score: {score:.4f}")

                if score < best_score:
                    best_score = score
                    best_params = (n_mfcc, n_fft, hop_length, n_mels)

print("\nBest hyperparameters:")
print(f"n_mfcc={best_params[0]}, n_fft={best_params[1]}, hop_length={best_params[2]}, n_mels={best_params[3]}")
print(f"Best score: {best_score:.4f}")

# Optionally, you can return the best parameters if needed
# return best_params
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchaudio
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from transformers import Wav2Vec2Processor
from jiwer import wer
import os

# Hyperparameters
n_mfcc = 30
n_fft = 1024
hop_length = 320
n_mels = 40
num_epochs = 10
hidden_size = 128
cnn_channels = 128
lstm_hidden_size = 256
num_layers = 3
learning_rate = 0.00001
num_classes = 45

# Load the transcripts dataframe
transcripts_df = pd.read_csv('/content/GenNote/train.csv')

# Define the dataset class
class AudioDataset(Dataset):
    def __init__(self, audio_files, transcripts, sample_rate=16000):
        self.audio_files = audio_files
        self.transcripts = transcripts
        self.sample_rate = sample_rate

    def __len__(self):
        return len(self.audio_files)

    def __getitem__(self, idx):
        file_path = self.audio_files[idx]
        transcript = self.transcripts[idx]
        audio, sample_rate = self.load_and_resample_audio(file_path, self.sample_rate)
        if audio is not None:
            mfcc = self.extract_mfcc(audio, sample_rate)
            return mfcc, transcript
        return None, None

    def load_and_resample_audio(self, file_path, target_sample_rate=16000):
        try:
            waveform, sample_rate = torchaudio.load(file_path)
            if sample_rate != target_sample_rate:
                waveform = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=target_sample_rate)(waveform)
            return waveform.squeeze(), target_sample_rate
        except Exception as e:
            print(f"Error loading {file_path}: {e}")
            return None, None

    def extract_mfcc(self, audio, sample_rate):
        mfcc_transform = torchaudio.transforms.MFCC(
            sample_rate=sample_rate,
            n_mfcc=n_mfcc,
            melkwargs={
                'n_mels': n_mels,
                'n_fft': n_fft,
                'hop_length': hop_length
            }
        )
        return mfcc_transform(audio)

def collate_fn(batch):
    batch = [item for item in batch if item[0] is not None]
    mfccs = [item[0] for item in batch]
    transcripts = [item[1] for item in batch]

    # Pad sequences to the maximum length in the batch
    max_length = max(mfcc.shape[1] for mfcc in mfccs)
    padded_mfccs = torch.zeros((len(mfccs), mfccs[0].shape[0], max_length))
    for i, mfcc in enumerate(mfccs):
        padded_mfccs[i, :, :mfcc.shape[1]] = mfcc

    # Create tensors for targets and lengths
    targets = [torch.tensor([ord(c) for c in t]) for t in transcripts]  # Convert transcripts to integer lists
    targets_padded = nn.utils.rnn.pad_sequence(targets, batch_first=True, padding_value=0)
    target_lengths = torch.tensor([len(t) for t in targets])

    input_lengths = torch.tensor([mfcc.shape[1] for mfcc in mfccs], dtype=torch.long)

    return padded_mfccs, targets_padded, input_lengths, target_lengths

class CNNLSTMCTCModel(nn.Module):
    def __init__(self, input_size, num_classes, cnn_channels, lstm_hidden_size, num_layers):
        super(CNNLSTMCTCModel, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv1d(input_size, cnn_channels, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv1d(cnn_channels, cnn_channels * 2, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
        )
        self.lstm = nn.LSTM(cnn_channels * 2, lstm_hidden_size, num_layers, batch_first=True, bidirectional=True)
        self.fc = nn.Linear(lstm_hidden_size * 2, num_classes)
        self.ctc_loss = nn.CTCLoss(blank=0, zero_infinity=True)

    def forward(self, x, targets, input_lengths, target_lengths):
        x = self.cnn(x)
        print(f"Shape after CNN: {x.shape}")

        # Update input_lengths after CNN
        cnn_output_lengths = input_lengths  # No max pooling
        x = x.permute(0, 2, 1)  # Change to (batch_size, time, features) for LSTM
        x, _ = self.lstm(x)
        print(f"Shape after LSTM: {x.shape}")
        x = self.fc(x)
        x = x.log_softmax(2)
        x = x.permute(1, 0, 2)  # Change to (time, batch_size, features) for CTC
        print(f"Shape before CTC Loss: {x.shape}")

        # Debug prints for lengths
        print(f"Input lengths: {cnn_output_lengths}")
        print(f"Target lengths: {target_lengths}")
        print(f"Targets: {targets}")

        # Ensure lengths are tensors
        if not isinstance(cnn_output_lengths, torch.Tensor):
            cnn_output_lengths = torch.tensor(cnn_output_lengths, dtype=torch.long)
        if not isinstance(target_lengths, torch.Tensor):
            target_lengths = torch.tensor(target_lengths, dtype=torch.long)

        loss = self.ctc_loss(x, targets, cnn_output_lengths, target_lengths)
        print(f"CTC Loss: {loss}")
        return loss, x

def train_model(model, optimizer, train_loader, num_epochs):
    model.train()
    for epoch in range(num_epochs):
        total_loss = 0.0
        for mfccs, targets, input_lengths, target_lengths in train_loader:
            mfccs, targets = mfccs.to('cuda'), targets.to('cuda')
            optimizer.zero_grad()
            loss, _ = model(mfccs, targets, input_lengths, target_lengths)
            if torch.isfinite(loss):
                loss.backward()
                optimizer.step()
                total_loss += loss.item()
            else:
                print(f"Skipping non-finite loss at batch with input_lengths {input_lengths} and target_lengths {target_lengths}")
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss:.4f}")

def evaluate_model(model, data_loader):
    model.eval()
    wer_scores = []
    processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
    with torch.no_grad():
        for mfccs, targets, input_lengths, target_lengths in data_loader:
            mfccs = mfccs.to('cuda')
            _, outputs = model(mfccs, targets, input_lengths, target_lengths)
            predicted_ids = torch.argmax(outputs, dim=-1)
            predicted_texts = processor.batch_decode(predicted_ids.cpu().numpy())
            for true_text, pred_text in zip(targets, predicted_texts):
                true_text = ''.join([chr(c) for c in true_text if c != 0])  # Convert back to string and remove padding
                wer_score = wer(true_text, pred_text)
                wer_scores.append(wer_score)
    avg_wer = sum(wer_scores) / len(wer_scores)
    print(f"Average WER: {avg_wer}")
    return avg_wer

def prepare_data(audio_dir, transcripts_df):
    audio_files = [os.path.join(audio_dir, f) for f in os.listdir(audio_dir) if f.endswith(".wav")]
    audio_file_ids = [os.path.splitext(f)[0] for f in os.listdir(audio_dir) if f.endswith(".wav")]
    filtered_df = transcripts_df[transcripts_df['encounter_id'].astype(str).isin(audio_file_ids)]
    audio_files = [os.path.join(audio_dir, f"{encounter_id}.wav") for encounter_id in filtered_df['encounter_id']]
    transcripts = filtered_df['dialogue'].tolist()
    return AudioDataset(audio_files, transcripts)

# Define your dataset, data loader, and other necessary components
audio_dir = "/content/ok"

# Prepare the dataset and dataloaders
dataset = prepare_data(audio_dir, transcripts_df)
train_loader = DataLoader(dataset, batch_size=2, shuffle=True, collate_fn=collate_fn)
eval_loader = DataLoader(dataset, batch_size=1, shuffle=False, collate_fn=collate_fn)

# Model: CNN + LSTM + CTC
model1 = CNNLSTMCTCModel(input_size=n_mfcc, num_classes=num_classes, cnn_channels=cnn_channels, lstm_hidden_size=lstm_hidden_size, num_layers=num_layers).to('cuda')
optimizer1 = optim.Adam(model1.parameters(), lr=learning_rate)

# Train and evaluate the model
train_model(model1, optimizer1, train_loader, num_epochs)
print("Model 1 evaluation:")
evaluate_model(model1, eval_loader)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from jiwer import wer
import os
import whisper

# Load the transcripts dataframe
transcripts_df = pd.read_csv('/content/GenNote/train.csv')


# Hyperparameters
num_epochs = 20
learning_rate = 0.001
num_classes = 128  # ASCII range

# Define the dataset class
class TextDataset(Dataset):
    def __init__(self, transcriptions_dict):
        self.keys = list(transcriptions_dict.keys())
        self.transcriptions = list(transcriptions_dict.values())
        self.transcripts = transcripts_df['dialogue'].tolist()

    def __len__(self):
        return len(self.transcriptions)

    def __getitem__(self, idx):
        key = self.keys[idx]
        transcription = self.transcriptions[idx]
        transcript = self.transcripts[idx]
        return key, transcription, transcript

def collate_fn(batch):
    keys = [item[0] for item in batch]
    transcriptions = [item[1] for item in batch]
    transcripts = [item[2] for item in batch]

    # Convert transcriptions and transcripts to ASCII values and pad them
    transcriptions_tensor = [torch.tensor([ord(c) for c in t]) for t in transcriptions]
    transcripts_tensor = [torch.tensor([ord(c) for c in t]) for t in transcripts]

    # Find the max length in the batch for padding
    max_length = max(max(len(t) for t in transcriptions_tensor), max(len(t) for t in transcripts_tensor))

    # Pad sequences to the max length
    transcriptions_padded = nn.utils.rnn.pad_sequence(transcriptions_tensor, batch_first=True, padding_value=0)
    transcripts_padded = nn.utils.rnn.pad_sequence(transcripts_tensor, batch_first=True, padding_value=0)

    transcriptions_padded = torch.nn.functional.pad(transcriptions_padded, (0, max_length - transcriptions_padded.size(1)), value=0)
    transcripts_padded = torch.nn.functional.pad(transcripts_padded, (0, max_length - transcripts_padded.size(1)), value=0)

    target_lengths = torch.tensor([len(t) for t in transcripts_tensor])

    return keys, transcriptions_padded, transcripts_padded, target_lengths

class TextMatchingModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim):
        super(TextMatchingModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)
        self.criterion = nn.CrossEntropyLoss(ignore_index=0)  # Ignoring padding index

    def forward(self, x, targets):
        x = self.embedding(x)
        x, _ = self.lstm(x)
        x = self.fc(x)  # Shape: (batch_size, seq_len, output_dim)

        # Reshape to (batch_size * seq_len, output_dim) for CrossEntropyLoss
        x = x.contiguous().view(-1, x.size(-1))  # Shape: (batch_size * seq_len, output_dim)
        targets = targets.contiguous().view(-1)  # Shape: (batch_size * seq_len)

        # Calculate the loss
        loss = self.criterion(x, targets)
        return loss, x

def train_model(model, optimizer, train_loader, num_epochs):
    model.train()
    for epoch in range(num_epochs):
        total_loss = 0.0
        for keys, transcriptions, targets, target_lengths in train_loader:
            optimizer.zero_grad()
            loss, _ = model(transcriptions, targets)
            if torch.isfinite(loss):
                loss.backward()
                optimizer.step()
                total_loss += loss.item()
            else:
                print(f"Skipping non-finite loss at batch with target_lengths {target_lengths}")
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss:.4f}")

def evaluate_model(model, data_loader):
    model.eval()
    results = []
    with torch.no_grad():
        for keys, transcriptions, targets, target_lengths in data_loader:
            _, outputs = model(transcriptions, targets)
            predicted_ids = torch.argmax(outputs, dim=-1)
            predicted_texts = ["".join([chr(i) for i in pred if i != 0]) for pred in predicted_ids]
            for key, true_text, pred_text in zip(keys, targets, predicted_texts):
                true_text = ''.join([chr(c) for c in true_text if c != 0])
                encounter_id = key.split('.')[0]  # Remove the .wav extension
                wer_score = wer(true_text, pred_text)
                results.append({
                    "encounter_id": encounter_id,
                    "dialogue": pred_text,
                    "wer": wer_score
                })
                print(f"True: {true_text} | Pred: {pred_text} | WER: {wer_score}")
    results_df = pd.DataFrame(results)
    avg_wer = results_df['wer'].mean()
    print(f"Average WER: {avg_wer}")
    return results_df



# Prepare the dataset and dataloaders
dataset = TextDataset(transcriptions_dict)
train_loader = DataLoader(dataset, batch_size=2, shuffle=True, collate_fn=collate_fn)
eval_loader = DataLoader(dataset, batch_size=1, shuffle=False, collate_fn=collate_fn)

# Model: Simple text matching model
embedding_dim = 64
hidden_dim = 128
output_dim = num_classes
model = TextMatchingModel(num_classes, embedding_dim, hidden_dim, output_dim)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Train and evaluate the model
train_model(model, optimizer, train_loader, num_epochs)
print("Model evaluation:")
results_df = evaluate_model(model, eval_loader)

# Display the results DataFrame
print(results_df)
```

```python
!pip install transformers torchaudio jiwer

import os
import torch
import torch.nn as nn
import torch.optim as optim
import torchaudio
import pandas as pd
import numpy as np
from transformers import Wav2Vec2Processor
from jiwer import wer

# Hyperparameters
n_mfcc = 13
n_fft = 400
hop_length = 160
n_mels = 23
hidden_size = 128
cnn_channels = 16
lstm_hidden_size = 32
num_layers = 1
d_model = 64
nhead = 1
num_transformer_layers = 1
learning_rate = 0.0001
gradient_clip_value = 1.0
weight_decay = 1e-5
batch_size = 2
num_epochs = 10

# Load the transcripts dataframe
transcripts_df = pd.read_csv('/content/GenNote/train.csv')

# Define the dataset class for multiple files
class AudioDataset(torch.utils.data.Dataset):
    def __init__(self, file_paths, transcripts, sample_rate=16000):
        self.file_paths = file_paths
        self.transcripts = transcripts
        self.sample_rate = sample_rate

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        audio_file = self.file_paths[idx]
        transcript = self.transcripts[idx]
        audio, sample_rate = self.load_and_resample_audio(audio_file, self.sample_rate)
        if audio is not None:
            mfcc = self.extract_mfcc(audio, sample_rate)
            mfcc = self.normalize_mfcc(mfcc)
            mfcc = torch.clamp(mfcc, -10, 10)
            transcript_tensor = torch.tensor([ord(c) for c in transcript])
            return mfcc, transcript_tensor
        return None, None

    def load_and_resample_audio(self, file_path, target_sample_rate=16000):
        try:
            waveform, sample_rate = torchaudio.load(file_path)
            if sample_rate != target_sample_rate:
                waveform = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=target_sample_rate)(waveform)
            return waveform.squeeze(), target_sample_rate
        except Exception as e:
            print(f"Error loading {file_path}: {e}")
            return None, None

    def extract_mfcc(self, audio, sample_rate):
        mfcc_transform = torchaudio.transforms.MFCC(
            sample_rate=sample_rate,
            n_mfcc=n_mfcc,
            melkwargs={
                'n_mels': n_mels,
                'n_fft': n_fft,
                'hop_length': hop_length
            }
        )
        return mfcc_transform(audio)

    def normalize_mfcc(self, mfcc):
        mean = mfcc.mean(dim=-1, keepdim=True)
        std = mfcc.std(dim=-1, keepdim=True)
        normalized_mfcc = (mfcc - mean) / (std + 1e-5)
        assert torch.isfinite(normalized_mfcc).all(), "Normalization resulted in non-finite values"
        return normalized_mfcc

class TransformerEncoderLayer(nn.Module):
    def __init__(self, d_model, nhead, dim_feedforward=256, dropout=0.1):
        super(TransformerEncoderLayer, self).__init__()
        self.self_attn = nn.MultiheadAttention(d_model, nhead, dropout=dropout)
        self.linear1 = nn.Linear(d_model, dim_feedforward)
        self.dropout = nn.Dropout(dropout)
        self.linear2 = nn.Linear(dim_feedforward, d_model)
        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)
        self.dropout1 = nn.Dropout(dropout)
        self.dropout2 = nn.Dropout(dropout)
        self.activation = nn.ReLU()

    def forward(self, src):
        src2 = self.self_attn(src, src, src)[0]
        src = src + self.dropout1(src2)
        src = self.norm1(src)
        src2 = self.linear2(self.dropout(self.activation(self.linear1(src))))
        src = src + self.dropout2(src2)
        src = self.norm2(src)
        return src

class CNNLSTMTransformerCTCModel(nn.Module):
    def __init__(self, input_size, num_classes, cnn_channels, lstm_hidden_size, num_layers, d_model, nhead, num_transformer_layers):
        super(CNNLSTMTransformerCTCModel, self).__init__()
        # CNN Layers
        self.cnn = nn.Sequential(
            nn.Conv1d(input_size, cnn_channels, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Conv1d(cnn_channels, cnn_channels * 2, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.Dropout(0.1)  # Dropout for regularization
        )
        # LSTM Layer
        self.lstm = nn.LSTM(cnn_channels * 2, lstm_hidden_size, num_layers, batch_first=True, bidirectional=True, dropout=0.1)
        # Transformer Encoder Layers
        self.transformer_encoder_layers = nn.ModuleList([
            TransformerEncoderLayer(d_model, nhead) for _ in range(num_transformer_layers)
        ])
        # Fully Connected Layer
        self.fc = nn.Linear(lstm_hidden_size * 2, num_classes)
        # CTC Loss
        self.ctc_loss = nn.CTCLoss(blank=0, zero_infinity=True)

    def forward(self, x, targets, input_lengths, target_lengths):
        if not torch.isfinite(x).all():
            print("Non-finite input detected before CNN")

        x = self.cnn(x)
        if not torch.isfinite(x).all():
            print("Non-finite values detected after CNN")
            x = torch.nan_to_num(x)  # Replace NaNs and Infs with 0

        x = x.permute(0, 2, 1)  # Change to (batch_size, time, features) for LSTM

        x, _ = self.lstm(x)
        if not torch.isfinite(x).all():
            print("Non-finite values detected after LSTM")
            x = torch.nan_to_num(x)  # Replace NaNs and Infs with 0

        x = x.permute(1, 0, 2)  # Change to (time, batch_size, features) for Transformer
        for layer in self.transformer_encoder_layers:
            x = layer(x)
        x = x.permute(1, 0, 2)  # Change back to (batch_size, time, features)

        x = self.fc(x)
        if not torch.isfinite(x).all():
            print("Non-finite values detected after Fully Connected layer")
            x = torch.nan_to_num(x)  # Replace NaNs and Infs with 0

        x = x.log_softmax(2)
        if not torch.isfinite(x).all():
            print("Non-finite values detected after log_softmax")
            x = torch.nan_to_num(x)  # Replace NaNs and Infs with 0

        x = x.permute(1, 0, 2)  # Change to (time, batch_size, features) for CTC

        if input_lengths.min() < target_lengths.max():
            raise ValueError("Input lengths after down-sampling are shorter than the target lengths.")

        x = x.to(torch.float32)
        targets = targets.to(torch.int32)
        input_lengths = input_lengths.to(torch.int32)
        target_lengths = target_lengths.to(torch.int32)
        loss = self.ctc_loss(x, targets, input_lengths, target_lengths)
        if not torch.isfinite(loss).all():
            print("Non-finite CTC loss detected")
            loss = torch.nan_to_num(loss)
        return loss, x

def collate_fn(batch):
    batch = [item for item in batch if item[0] is not None and item[1] is not None]
    mfccs = [item[0].transpose(0, 1) for item in batch]  # Transpose to (features, time) for padding
    transcripts = [item[1] for item in batch]

    mfccs = torch.nn.utils.rnn.pad_sequence(mfccs, batch_first=True, padding_value=0).transpose(1, 2)  # Transpose back to (batch, time, features)
    input_lengths = torch.tensor([mfcc.size(1) for mfcc in mfccs])
    transcripts = torch.nn.utils.rnn.pad_sequence(transcripts, batch_first=True, padding_value=0)
    target_lengths = torch.tensor([len(transcript) for transcript in transcripts])

    return mfccs, transcripts, input_lengths, target_lengths

def train_model(model, optimizer, dataloader, num_epochs):
    device = torch.device('cpu')  # Use CPU to avoid GPU memory issues
    model = model.to(device)
    model.train()

    for epoch in range(num_epochs):
        for batch in dataloader:
            mfccs, targets, input_lengths, target_lengths = batch
            mfccs, targets, input_lengths, target_lengths = mfccs.to(device), targets.to(device), input_lengths.to(device), target_lengths.to(device)

            optimizer.zero_grad()
            loss, _ = model(mfccs, targets, input_lengths, target_lengths)
            if not torch.isfinite(loss).all():
                print(f"Found non-finite loss at epoch {epoch + 1}")
                continue
            print(f"Epoch {epoch + 1} Loss: {loss.item()}")
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), gradient_clip_value)
            optimizer.step()
            print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

def evaluate_model(model, dataset):
    device = torch.device('cpu')  # Use CPU to avoid GPU memory issues
    model.eval()
    processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
    mfccs, targets = dataset.mfcc.unsqueeze(0).to(device), dataset.transcript_tensor.unsqueeze(0).to(device)
    input_lengths = torch.tensor([mfccs.size(2)], dtype=torch.long).to(device)
    target_lengths = torch.tensor([targets.size(1)], dtype=torch.long).to(device)

    with torch.no_grad():
        loss, outputs = model(mfccs, targets, input_lengths, target_lengths)
        outputs = outputs.permute(1, 0, 2)
        predicted_ids = torch.argmax(outputs, dim=-1)
        predicted_texts = processor.batch_decode(predicted_ids.cpu(), skip_special_tokens=True)

        for true_text, pred_text in zip(targets, predicted_texts):
            true_text = ''.join([chr(c) for c in true_text if c != 0])
            pred_text = pred_text.replace(' ', '')
            wer_score = wer(true_text, pred_text)
            cer_score = cer(true_text, pred_text)
            print(f"WER: {wer_score}, CER: {cer_score}")

def cer(true_text, pred_text):
    true_text = list(true_text)
    pred_text = list(pred_text)
    errors = edit_distance(true_text, pred_text)
    return errors / len(true_text)

def edit_distance(s1, s2):
    if len(s1) < len(s2):
        return edit_distance(s2, s1)

    if len(s2) == 0:
        return len(s1)

    previous_row = range(len(s2) + 1)
    for i, c1 in enumerate(s1):
        current_row = [i + 1]
        for j, c2 in enumerate(s2):
            insertions = previous_row[j + 1] + 1
            deletions = current_row[j] + 1
            substitutions = previous_row[j] + (c1 != c2)
            current_row.append(min(insertions, deletions, substitutions))
        previous_row = current_row

    return previous_row[-1]

# Prepare training datasets
train_file_paths = [
    "/content/GenNote/Wav_Recordings/D2N005.wav",
    "/content/GenNote/Wav_Recordings/D2N017.wav",
    "/content/GenNote/Wav_Recordings/D2N012.wav"
]
train_transcripts = [
    transcripts_df[transcripts_df['encounter_id'] == 'D2N005']['dialogue'].iloc[0],
    transcripts_df[transcripts_df['encounter_id'] == 'D2N017']['dialogue'].iloc[0],
    transcripts_df[transcripts_df['encounter_id'] == 'D2N012']['dialogue'].iloc[0]
]

train_dataset = AudioDataset(train_file_paths, train_transcripts)
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True, collate_fn=collate_fn)

# Prepare the single test dataset
test_audio_file = "/content/GenNote/Wav_Recordings/D2N011.wav"
test_transcript = transcripts_df[transcripts_df['encounter_id'] == 'D2N011']['dialogue'].iloc[0]

# Define the dataset class for a single file
class SingleFileDataset:
    def __init__(self, audio_file, transcript, sample_rate=16000):
        self.audio_file = audio_file
        self.transcript = transcript
        self.sample_rate = sample_rate
        self.mfcc, self.transcript_tensor = self.preprocess()

    def preprocess(self):
        audio, sample_rate = self.load_and_resample_audio(self.audio_file, self.sample_rate)
        if audio is not None:
            mfcc = self.extract_mfcc(audio, sample_rate)
            mfcc = self.normalize_mfcc(mfcc)
            mfcc = torch.clamp(mfcc, -10, 10)
            transcript_tensor = torch.tensor([ord(c) for c in self.transcript])
            return mfcc, transcript_tensor
        return None, None

    def load_and_resample_audio(self, file_path, target_sample_rate=16000):
        try:
            waveform, sample_rate = torchaudio.load(file_path)
            if sample_rate != target_sample_rate:
                waveform = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=target_sample_rate)(waveform)
            return waveform.squeeze(), target_sample_rate
        except Exception as e:
            print(f"Error loading {file_path}: {e}")
            return None, None

    def extract_mfcc(self, audio, sample_rate):
        mfcc_transform = torchaudio.transforms.MFCC(
            sample_rate=sample_rate,
            n_mfcc=n_mfcc,
            melkwargs={
                'n_mels': n_mels,
                'n_fft': n_fft,
                'hop_length': hop_length
            }
        )
        return mfcc_transform(audio)

    def normalize_mfcc(self, mfcc):
        mean = mfcc.mean(dim=-1, keepdim=True)
        std = mfcc.std(dim=-1, keepdim=True)
        normalized_mfcc = (mfcc - mean) / (std + 1e-5)
        assert torch.isfinite(normalized_mfcc).all(), "Normalization resulted in non-finite values"
        return normalized_mfcc

test_dataset = SingleFileDataset(test_audio_file, test_transcript)

# Create the model
num_classes = len(set("".join(pd.read_csv('/content/GenNote/train.csv')['dialogue']))) + 1
model = CNNLSTMTransformerCTCModel(
    input_size=n_mfcc,
    num_classes=num_classes,
    cnn_channels=cnn_channels,
    lstm_hidden_size=lstm_hidden_size,
    num_layers=num_layers,
    d_model=d_model,
    nhead=nhead,
    num_transformer_layers=num_transformer_layers
)
optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=weight_decay)

# Train the model
train_model(model, optimizer, train_dataloader, num_epochs=num_epochs)

# Evaluate the model
print("Model evaluation:")
evaluate_model(model, test_dataset)
```
